```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from itertools import permutations
```

```python
def generateId(df, section='A'):
  n1 = np.array(df.Nombre1)
  a1 = np.array(df.Apellido1)
  id = []
  for i in range(len(n1)):
    id.append(section+str(i)+str(n1[i][0])+str(a1[i][0]))
  return id
```

# Seccción A

```python
section_A = {'Nombre1': ['David', 'Carlos', 'Julio', 'Carmen'],
             'Nombre2': ['Alfonso', 'Alberto', 'Sebastian', 'Dina'],
             'Apellido1': ['Lavan', 'Montenegro', 'Perez', 'Ortega'],
             'Apellido2': ['Quiroz', 'Vasquez', 'Vasques', 'Cabrejos']
     }
dfA= pd.DataFrame(data=section_A)
dfA
```

```python
perm_nombres = list(permutations(pd.concat([dfA['Nombre1'], dfA['Nombre2']]), 2))
perm_apellidos = list(permutations(pd.concat([dfA['Apellido1'], dfA['Apellido2']]), 2))
matrix_section_A = [[nombre[0], nombre[1], apellido[0], apellido[1]] for nombre in perm_nombres for apellido in perm_apellidos]
df_section_A = pd.DataFrame(data=matrix_section_A, columns=['Nombre1', 'Nombre2', 'Apellido1', 'Apellido2'])
df_section_A['id'] = generateId(df_section_A)
df_section_A.set_index('id',inplace=True)
df_section_A.drop_duplicates(inplace=True)
df_section_A
```

# Seccción B

```python
section_B = {'Nombre1': ['Lola', 'Pastor', 'Julia', 'Maria'],
             'Nombre2': ['Zoraida', 'Ausberto', 'Liana', 'Eri'],
             'Apellido1': ['Franco', 'Rojas', 'Collante', 'Cerron'],
             'Apellido2': ['Honores', 'Diaz', 'Castillo', 'Vela']
     }
dfB = pd.DataFrame(data=section_B)
dfB
```

```python
perm_nombres = list(permutations(pd.concat([dfB['Nombre1'], dfB['Nombre2']]), 2))
perm_apellidos = list(permutations(pd.concat([dfB['Apellido1'], dfB['Apellido2']]), 2))
matrix_section_A = [[nombre[0], nombre[1], apellido[0], apellido[1]] for nombre in perm_nombres for apellido in perm_apellidos]
df_section_B = pd.DataFrame(data=matrix_section_A, columns=['Nombre1', 'Nombre2', 'Apellido1', 'Apellido2'])
df_section_B['id'] = generateId(df_section_B, 'B')
df_section_B.set_index('id',inplace=True)
df_section_B.drop_duplicates(inplace=True)
df_section_B
```

# Seccción C

```python
section_C = {'Nombre1': ['Erika', 'Leonardo', 'Fiorella', 'Jesus'],
             'Nombre2': ['Soledad', 'Angelo', 'Mayte', 'Manuel'],
             'Apellido1': ['Remon', 'Quispe', 'Silvera', 'Delgado'],
             'Apellido2': ['Ancassi', 'Rio', 'LLufire', 'Pachas']
     }
dfC = pd.DataFrame(data=section_C)
dfC
```

```python
perm_nombres = list(permutations(pd.concat([dfC['Nombre1'], dfC['Nombre2']]), 2))
perm_apellidos = list(permutations(pd.concat([dfC['Apellido1'], dfC['Apellido2']]), 2))
matrix_section_A = [[nombre[0], nombre[1], apellido[0], apellido[1]] for nombre in perm_nombres for apellido in perm_apellidos]
df_section_C = pd.DataFrame(data=matrix_section_A, columns=['Nombre1', 'Nombre2', 'Apellido1', 'Apellido2'])
df_section_C['id'] = generateId(df_section_A, 'C')
df_section_C.set_index('id',inplace=True)
df_section_C.drop_duplicates(inplace=True)
df_section_C
```

# Todas las secciones

```python
df_all = pd.concat([df_section_A, df_section_B, df_section_C])
df_all
```

```python
df_all_sample = df_all.sample(frac=0.75, replace=False, random_state=1)
df_all_sample
```

```python
count_A = len(df_all_sample[df_all_sample.index.str.startswith("A")].index)
count_B = len(df_all_sample[df_all_sample.index.str.startswith("B")].index)
count_C = len(df_all_sample[df_all_sample.index.str.startswith("C")].index)
```

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')


fig, ax = plt.subplots()
p = ax.bar(['A', 'B', 'C'], [count_A,count_B, count_C])
ax.set_xticks(np.arange(3))
ax.set_ylim(0, 3000)
ax.set_xticklabels(['A', 'B', 'C'])
autolabel(p)
plt.tight_layout()
plt.show()
```
